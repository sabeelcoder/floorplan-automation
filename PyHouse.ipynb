```python
import petl
import MySQLdb
import math

def rotate(row,*args):
 
    #Rotate a point counterclockwise by a given angle around a given origin.
    
    #Then the origins each point references are reset

    #The angle should be given in radians.
    
    #The input would be two values to a point

    ox, oy = args[0]
    angle =  math.radians(args[1])
    mode=args[2]
    if mode=="usp":
        px, py = row['UnitStartPointX'],row['UnitStartPointY']
    if mode=="uep":
        px, py = row['UnitEndPointX'],row['UnitEndPointY']
    if mode=="sp":        
        px, py = row['StartPointX'],row['StartPointY']  
    if mode=="ep":          
        px, py = row['EndPointX'],row['EndPointY']  
    if mode=="cen":  
        px, py = row['CentroidX'],row['CentroidY']       
    if mode=="ucen":  
        px, py = row['UnitCentroidX'],row['UnitCentroidY']
    
    qx = (ox + math.cos(angle) * (px - ox) - math.sin(angle) * (py - oy))- ox
    qy = (oy + math.sin(angle) * (px - ox) + math.cos(angle) * (py - oy))- oy
 
    return qx, qy


from petl import look, fromdb, select, selecteq, convert, fieldmap

connection = MySQLdb.connect(host="localhost",    # connection to the local host
                     user="root",         
                     passwd="samwillis18", 
                     db="mydb") 
connection.cursor().execute('SET SQL_MODE=ANSI_QUOTES')

unitdata = fromdb(connection, 'select * from unitdata')
unitboundaries = fromdb(connection, 'select * from unitboundaries')
roomdata = fromdb(connection, 'select * from roomdata')
roomboundaries = fromdb(connection, 'select * from roomboundaries')

look(unitdata)
```

```python
def order(points):
    ordered=[]
    ordered.append(points.pop(0))
    for i in range(1,len(points)+1):
        
        close=closest(ordered[i-1],points)
        ordered.append(close)
        points.remove(close)
        
    return ordered
    
```
